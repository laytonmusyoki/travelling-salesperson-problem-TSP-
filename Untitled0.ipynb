```python
import networkx as nx

def mst_heuristic(graph):
    """
    Calculate the MST cost for a given graph as an estimate for TSP cost.

    Parameters:
    graph (nx.Graph): A complete graph where each edge has a weight representing the distance between cities.

    Returns:
    float: The cost of the MST.
    """
    mst = nx.minimum_spanning_tree(graph, algorithm='kruskal')
    mst_cost = mst.size(weight='weight')
    return mst_cost

# Example usage
if __name__ == "__main__":
    # Create a complete graph with distances as edge weights
    G = nx.Graph()
    edges = [
        (0, 1, 10),
        (0, 2, 15),
        (0, 3, 20),
        (1, 2, 35),
        (1, 3, 25),
        (2, 3, 30)
    ]
    G.add_weighted_edges_from(edges)

    # Compute MST heuristic for the TSP
    mst_cost = mst_heuristic(G)
    print(f"The MST cost is: {mst_cost}")
```

```python
import networkx as nx
import math

def straight_line_distance_heuristic(cities):
    """
    Calculate the straight-line distance heuristic for a given set of cities.

    Parameters:
    cities (list of tuples): List of (x, y) coordinates for each city.

    Returns:
    float: The estimated TSP cost using the straight-line distance heuristic.
    """
    total_distance = 0
    num_cities = len(cities)
    for i in range(num_cities):
        x1, y1 = cities[i]
        x2, y2 = cities[(i + 1) % num_cities]  # Wrap around to the first city
        total_distance += math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return total_distance

def mst_heuristic(cities):
    """
    Calculate the MST cost for a given set of cities as an estimate for TSP cost.

    Parameters:
    cities (list of tuples): List of (x, y) coordinates for each city.

    Returns:
    float: The cost of the MST.
    """
    G = nx.Graph()
    num_cities = len(cities)
    for i in range(num_cities):
        for j in range(i + 1, num_cities):
            x1, y1 = cities[i]
            x2, y2 = cities[j]
            distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
            G.add_edge(i, j, weight=distance)

    mst = nx.minimum_spanning_tree(G, algorithm='kruskal')
    mst_cost = mst.size(weight='weight')
    return mst_cost

# Example usage
if __name__ == "__main__":
    # Define cities as (x, y) coordinates
    cities = [
        (0, 0),  # City A
        (0, 1),  # City B
        (1, 0),  # City C
        (1, 1)   # City D
    ]

    straight_line_distance = straight_line_distance_heuristic(cities)
    mst_cost = mst_heuristic(cities)

    print(f"The straight-line distance heuristic is: {straight_line_distance}")
    print(f"The MST cost is: {mst_cost}")

    if mst_cost >= straight_line_distance:
        print("MST heuristic dominates the straight-line distance heuristic.")
    else:
        print("Straight-line distance heuristic dominates the MST heuristic.")
```

```python
import random
import math
import heapq
import networkx as nx
import matplotlib.pyplot as plt

def generate_tsp_instance(num_cities):
    """
    Generate a TSP instance with cities represented by random points in the unit square.

    Parameters:
    num_cities (int): The number of cities.

    Returns:
    list of tuples: A list of (x, y) coordinates representing the cities.
    """
    cities = [(random.random(), random.random()) for _ in range(num_cities)]
    return cities

def plot_cities(cities):
    """
    Plot the cities on a unit square.

    Parameters:
    cities (list of tuples): A list of (x, y) coordinates representing the cities.
    """
    x, y = zip(*cities)
    plt.scatter(x, y)
    plt.title("TSP Cities in the Unit Square")
    plt.xlabel("x-coordinate")
    plt.ylabel("y-coordinate")
    plt.show()

def euclidean_distance(city1, city2):
    """
    Calculate the Euclidean distance between two cities.

    Parameters:
    city1, city2 (tuple): Coordinates of the two cities.

    Returns:
    float: Euclidean distance between the cities.
    """
    return math.sqrt((city1[0] - city2[0]) ** 2 + (city1[1] - city2[1]) ** 2)

def mst_heuristic(cities, current_path):
    """
    Calculate the MST cost for the remaining cities using Prim's algorithm.

    Parameters:
    cities (list of tuples): List of all cities' coordinates.
    current_path (list of int): List of indices of cities in the current path.

    Returns:
    float: The cost of the MST for the remaining cities.
    """
    remaining_cities = set(range(len(cities))) - set(current_path)
    if not remaining_cities:
        return 0

    G = nx.Graph()
    for i in remaining_cities:
        for j in remaining_cities:
            if i != j:
                G.add_edge(i, j, weight=euclidean_distance(cities[i], cities[j]))

    mst = nx.minimum_spanning_tree(G, algorithm='prim')
    return mst.size(weight='weight')

def a_star_tsp(cities):
    """
    Solve the TSP using A* search with the MST heuristic.

    Parameters:
    cities (list of tuples): List of all cities' coordinates.

    Returns:
    list of int: The order of cities in the optimal TSP path.
    float: The total cost of the optimal TSP path.
    """
    num_cities = len(cities)
    start_city = 0
    open_list = []
    heapq.heappush(open_list, (0, start_city, [start_city], 0))
    best_cost = float('inf')
    best_path = None

    while open_list:
        _, current_city, current_path, current_cost = heapq.heappop(open_list)

        if len(current_path) == num_cities:
            total_cost = current_cost + euclidean_distance(cities[current_path[-1]], cities[start_city])
            if total_cost < best_cost:
                best_cost = total_cost
                best_path = current_path
            continue

        for next_city in range(num_cities):
            if next_city not in current_path:
                new_path = current_path + [next_city]
                new_cost = current_cost + euclidean_distance(cities[current_city], cities[next_city])
                heuristic_cost = mst_heuristic(cities, new_path)
                total_estimated_cost = new_cost + heuristic_cost
                if total_estimated_cost < best_cost:
                    heapq.heappush(open_list, (total_estimated_cost, next_city, new_path, new_cost))

    return best_path + [start_city], best_cost

# Example usage
if __name__ == "__main__":
    num_cities = 10
    cities = generate_tsp_instance(num_cities)
    plot_cities(cities)

    optimal_path, optimal_cost = a_star_tsp(cities)
    print("Optimal TSP Path:", optimal_path)
    print("Optimal TSP Cost:", optimal_cost)

    # Plot the optimal path
    optimal_path_coords = [cities[i] for i in optimal_path]
    x, y = zip(*optimal_path_coords)
    plt.plot(x, y, marker='o')
    plt.title("Optimal TSP Path")
    plt.xlabel("x-coordinate")
    plt.ylabel("y-coordinate")
    plt.show()
```
